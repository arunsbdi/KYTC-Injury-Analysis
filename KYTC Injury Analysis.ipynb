```python
import matplotlib.pyplot as plt
import pandas as pd


# Convert the entire 'Total Incurred' column to strings
df['Total Incurred'] = df['Total Incurred'].astype(str)

# Replace non-numeric characters and convert to float
df['Total Incurred'] = pd.to_numeric(df['Total Incurred'].str.replace('[^\d.]', ''), errors='coerce')

# Filter the DataFrame for years between 1992 and 2022 and convert 'Policy Year' to integer
df_filtered = df[df['Policy Year'].between(1992, 2022)]
df_filtered['Policy Year'] = df_filtered['Policy Year'].astype(int)

#Group by 'Policy year' and count the number of entries on any column
yearly_counts = df_filtered.groupby('Policy Year')['Total Incurred'].count()

# Group by 'Policy Year' and sum 'Total Incurred' for each year
total_incurred_by_year = df_filtered.groupby('Policy Year')['Total Incurred'].sum()

# Plotting with a secondary y-axis for 'Total Incurred'
fig, ax1 = plt.subplots(figsize=(15, 6))

# Assuming 'yearly_counts' contains the count of cases per year
ax1.plot(yearly_counts.index, yearly_counts.values, color='blue', marker='o', label='Number of Cases')
ax1.set_xlabel('Year')
ax1.set_ylabel('Number of Cases', color='blue')
ax1.tick_params(axis='y', labelcolor='blue')

# Secondary y-axis for 'Total Incurred'
ax2 = ax1.twinx()
ax2.plot(total_incurred_by_year.index, total_incurred_by_year.values, color='green', marker='x', label='Total Incurred')
ax2.set_ylabel('Total Incurred', color='green')
ax2.tick_params(axis='y', labelcolor='green')

# Add title and grid
plt.title('Number of Cases and Total Incurred per Year (1992-2022)')
ax1.grid(True)

# Set x-axis ticks to show every year from 1992 to 2023
plt.xticks(range(1992, 2023), rotation=90)

# Show the legends
ax1.legend(loc='upper left')
ax2.legend(loc='upper right')

# Adjust layout
plt.tight_layout()
plt.show()
```

```python
import numpy as np

# Define the unique districts ensuring 'District 1' starts first
unique_districts = sorted(df_filtered['District'].dropna().unique(), key=lambda x: (x != 'CO', x))

# Define a color palette with a distinct color for each district
colors = plt.cm.viridis(np.linspace(0, 1, len(unique_districts)))

# global maximum number of cases to set a consistent y-axis
global_max_cases = max(df_filtered.groupby(['District', 'Policy Year']).size())

# Create a figure and a grid of subplots
num_rows = (len(unique_districts) + 2) // 3
fig, axs = plt.subplots(num_rows, 3, figsize=(20, num_rows * 5), constrained_layout=True)
axs = axs.flatten()

# Loop over each unique district and plot the data
for i, (district, color) in enumerate(zip(unique_districts, colors)):
    # Filter the DataFrame for the current district
    district_df = df_filtered[df_filtered['District'] == district]

    # Group by 'Policy Year' and count the number of cases
    district_yearly_counts = district_df.groupby('Policy Year').size()

    # Plot the district data on the appropriate subplot
    ax = axs[i]
    ax.plot(district_yearly_counts.index, district_yearly_counts, marker='o', linestyle='-', color=color, label=district)
    ax.set_title(district)
    ax.set_ylim(0, global_max_cases)  # Set the y-axis limit
    ax.set_xlabel('Year')
    ax.set_ylabel('Number of Cases')
    ax.legend()
    ax.grid(True)

# Hide any unused subplots
for j in range(i + 1, len(axs)):
    axs[j].axis('off')

# Show the plot
plt.show()
```

```python
from scipy.stats import chi2_contingency
import pandas as pd
import numpy as np
import io

# Read the file
data = pd.read_excel(io.BytesIO(uploaded['KYTC_Monthly 1-31-23[59] (1).xlsx']))

# Filter the data for the years 1992 to 2022
data_filtered = data[(data['Policy Year'] >= 1992) & (data['Policy Year'] <= 2022)]

# Remove rows with missing 'Cause Code' or 'District' values
data_clean = data_filtered.dropna(subset=['Cause Code', 'District'])

# Create a contingency table for 'District' and 'Cause Code'
contingency_table = pd.crosstab(data_clean['District'], data_clean['Cause Code'])

# Perform a Chi-Square Test of Independence
chi2_stat, p_val, dof, ex = chi2_contingency(contingency_table)

# Output the results of the chi-square test
print(f"Chi2 Stat: {chi2_stat}, P-value: {p_val}")

# Calculate standardized residuals
residuals = (contingency_table - ex) / np.sqrt(ex)
residuals_df = pd.DataFrame(residuals, index=contingency_table.index, columns=contingency_table.columns)

# Create an empty list to store DataFrames
frames = []

# Iterate over districts to find max residuals
for district in residuals_df.index:
    max_residual_cause_code = residuals_df.loc[district].abs().idxmax()
    max_residual_value = residuals_df.loc[district].abs().max()

    # Create a DataFrame for each result
    result_df = pd.DataFrame({'District': [district],
                              'Cause Code': [max_residual_cause_code],
                              'Standardized Residual': [max_residual_value]})

    # Append the DataFrame to the list
    frames.append(result_df)

# Concatenate the list of DataFrames into a single DataFrame
results_df = pd.concat(frames, ignore_index=True)

# Print the results in tabular form
results_df
```
